```python
filename = "elsysers.jpg"
result = extract_faces(open_image(filename))
display(Image.open(filename))
i = 0
for face in result:
    display(Image.fromarray(face))
    Image.fromarray(face).save(f'./out/{i}.jpg')
    i += 1
```
